Aim: To draw Gaussian peaks for $^{10}$B, $^{10}$Be and $^{9}$Be (from $^{9}$Be+H molecule). We would like to see how the spectrum changes if you change the foil thickness. 

Below for a 3um thick Havar foil. Ion species is 10B. Ebeam = 11 MeV 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
from scipy import signal

def gauss_function(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))

fp = 'Hist10BHavar3um.txt'
sim_data = pd.read_csv(fp, delim_whitespace=True, header=None)

sim_data.columns = ['energy','counts']

sim_data.head()

sim_data['energy']=sim_data['energy']/1e6

# append from pamdas to add values 

# sim_data.plot("energy","counts")

## create gaussian kernel ##
sigma=0.20399069
energy=sim_data['energy']
dx=(max(energy)-min(energy))/len(energy)
gaussian_x=np.arange(-3*sigma, 3*sigma, dx)
gaussian = np.exp(-(gaussian_x/sigma)**2/2)

sig=sim_data['counts']


filtered = signal.convolve(sig, gaussian, mode='same')/sum(gaussian)
# print(sum(sig), sum(filtered))

###########################################################
def gauss_function(x, a, x0, sigma):
    # Where a is the area of the Gaussian peak, x0 is the mean/centroid and sigma is the standard deviation. 
    return a*np.exp(-(x-x0)**2/(2*sigma**2))

# p0 = [1, mean, sigma]
popt, pcov = curve_fit(gauss_function, energy, filtered, p0 = [1, 4.0, 0.4])
# popt gives us the optimised values of a, x0, sigma

# print(popt)

############################################################

# print(len(energy), len(filtered), len(gaussian))
sim_data.plot("energy","counts",  label="Srim Simulation")
ax = plt.gca()
ax.set_xlabel("Energy, MeV")
ax.set_ylabel("Counts")
ax.set(xlim=(3, 5))
plt.plot(energy,filtered, label="Gaussian convoluted")
#gauss_function(x values !!, *popt)
plt.plot(energy, gauss_function(energy, *popt), label="Fitted Gaussian")
plt.legend()
plt.show()

print("The fitted gaussian parameters are: ", popt)
```
