```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

fp = 'Au_10B.mpa'
spectrum = pd.read_csv(fp, sep=',', skiprows=221)
spectrum.plot()
ax = plt.gca()
ax.set_xlabel("Channel number")
ax.set_ylabel("Counts")
ax.set(xlim=(0, 3000))
ax.legend().remove()
```
